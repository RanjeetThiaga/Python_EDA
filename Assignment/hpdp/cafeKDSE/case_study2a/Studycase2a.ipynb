```python
import sweetviz as sv
import pandas as pd
import IPython
```

## Importing the dataset


> This dataset pertains to the demography of vaccine recipients in terms of risk factors. It involves the observation of three groups: individuals with comorbidities, persons with disabilities (OKU), and healthcare workers. Each group records the total number of recipients for the partial dose, full dose, booster1, and booster2. Additionally, the dataset includes information about the vaccine center locations in each state and country, along with the date of vaccination for each group.


```python
df = pd.read_csv('https://raw.githubusercontent.com/MoH-Malaysia/covid19-public/main/vaccination/vax_demog_highrisk.csv')
df
```



> The dataset has been downloaded and extracted. Now, we can know the column of the dataset by using info() and describe() statements.

## Exploratory Analysis and Visualization



> In this study case, the EDA used will be SweetViz that have been installed and imported beforehand. Next, we just need to analyze our dataframe and use show_html to generate the report. If web browser not pop up, use the display.HTML() from IPython library to display the result in the Colab.





```python
report = sv.analyze(df)
```

```python
report.show_html('sweetviz_report.html')
```

```python
IPython.display.HTML('sweetviz_report.html')
```


> The EDA also can do comparison by analyzing two dataframes like in this example, the recent dartaframe will be splitted into new dataframe of states from peninsular and from East Malaysia.



```python
peninsular_states = ['Johor', 'Kedah', 'Kelantan', 'Melaka', 'Negeri Sembilan', 'Pahang', 'Perak', 'Perlis', 'Selangor', 'Terengganu']
east_states = ['Sabah', 'Sarawak']
```

```python
peninsular_condition = df['state'].isin(peninsular_states)
df_peninsular_states = df[peninsular_condition]
df_east_states = df[df['state'].isin(east_states) & ~peninsular_condition]
```

```python
report_compare = sv.compare([df_peninsular_states, 'Peninsular States Dataset'], [df_east_states, 'East States Dataset'])
report_compare.show_html('sweetviz_report_compare.html')
```

### Pros and Cons Analysis for SweetViz:

#### SweetViz:

**Pros:**
1. SweetViz only consisted of several lines or code for analysis execution.
2. Its ability to compare two datasets side by side is a valuable feature for understanding differences.
3. Its also displays complete and detail information for each column of EDA.
4. Results are presented in HTML reports, facilitating easy sharing and viewing.

**Cons:**
1. SweetViz reports are static HTML files, lacking the interactive elements which is only offer the association feature.
2. SweetViz only display the analysis in bar graph.

### Insights from Malaysian Dataset:

SweetViz performed admirably in analyzing the demographic and high-risk factors related to vaccine coverage in Malaysia. Its simplicity such as the bar graph is good for a quick yet detailed examination of the dataset, and the comparative analysis feature proved useful in understanding variations across different demographic groups such as for the state in peninsular or in East Malaysia.

### Conclusion:

Finally, SweetViz is notable for being simple to use and producing meaningful reports quickly. It is especially appropriate for those looking for a simple, automated method for exploratory data analysis—particularly in cases where comparison analysis is essential.
