##3. Data Preparation and Cleaning
For this section, we use the pandas library to read the CSV files into a dataframes we called our dataframe `resturants_data`, enabling us to begin working with our data. This involves validating the data to ensure there are no missing values, setting the correct data types, dropping unnecessary fields, and creating additional columns to enhance the meaningfulness of the data. The use of such libraries enhances the efficiency and accuracy of our data analysis process


```python
resturants_data = pd.read_csv("https://raw.githubusercontent.com/BakungaBronson/Trip-Advisor-Dataset/main/TripAdvisor_data_cleaned.csv")
resturants_data
```

```python

# We decided to drop of all rows whose date wasnt making sense like yesterday , 1 week , 2 weeks ago.
# we are interested in standard dates so that we don't have inconsistencies in our data .
# Apply the replacement
resturants_data['Dates'].replace('Reviewed ', '', regex=True, inplace=True)

# Convert to datetime and coerce errors to NaT
resturants_data['Dates'] = pd.to_datetime(resturants_data['Dates'], errors='coerce')

# Drop rows where 'Dates' is NaT a standard date
resturants_data = resturants_data.dropna(subset=['Dates'])
resturants_data
```

```python
columns_dataypes={
    "Author":str,
    "Title":str,
    "Review":str,
    "Rating":int,
    "Restaurant":str,
    "Location":str
}
resturants_data.astype(columns_dataypes)
resturants_data.dtypes
```

```python
# To keep our data more clean we have to remove those special charaters , trimming space and more
#The code is here
# Function to remove unwanted characters
def clean_column(column):
    column = column.str.strip()
    column = column.str.replace('[^a-zA-Z0-9\s]', '', regex=True)
    return column

# Apply the cleaning function to each column of type 'object'
for col in resturants_data.columns:
    if resturants_data[col].dtype == 'object':
        resturants_data.loc[:, col] = clean_column(resturants_data[col])
resturants_data.head(50)
```

For title and reviews we used an explatory analysis that involves correlation we used a tool called sentiment analysis to analyze both columns.Using the  TextBlob library provided by python .This library is a simple, user-friendly library for basic sentiment analysis. It assigns a polarity score ranging from -1 (negative) to +1 (positive).

```python
from textblob import TextBlob

# Function to calculate sentiment polarity
def calculate_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Apply sentiment analysis to the 'Title' and 'Review' columns
resturants_data['Title Sentiment'] = resturants_data['Title'].apply(calculate_sentiment)
resturants_data['Review Sentiment'] = resturants_data['Review'].apply(calculate_sentiment)
resturants_data['Title Sentiment'] ,resturants_data['Review Sentiment']
```

####Q3: How does the number of reviews vary across different months?

```python
import calendar
resturants_data['Dates'] = pd.to_datetime(resturants_data['Dates'])
resturants_data['Month'] = resturants_data['Dates'].dt.month

# Convert month numbers to month names
resturants_data['Month Name'] = resturants_data['Month'].apply(lambda x: calendar.month_name[x])

monthly_reviews = resturants_data.groupby('Month Name').size()

# Reorder the data by month order
ordered_months = [calendar.month_name[i] for i in range(1, 13)]
monthly_reviews = monthly_reviews.reindex(ordered_months)
monthly_reviews
```

```python
monthly_reviews.plot(kind='bar', color='coral')

plt.title('Number of Reviews per Month (2009-2019)')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)
plt.show()
```


####Q5: Which locations have the highest concentration of top-rated restaurants?

```python
# threshold for top-rated
top_rated_threshold = 3.5

# Filter the dataset for top-rated restaurants
top_rated_restaurants = resturants_data[resturants_data['Rating'] >= top_rated_threshold]

# Count the number of top-rated restaurants in each location
top_rated_count_by_location = top_rated_restaurants.groupby('Location').size()

# Sort the counts in descending order
top_rated_count_by_location_sorted = top_rated_count_by_location.sort_values(ascending=False)

# top N locations like here n=10
top_locations = top_rated_count_by_location_sorted.head(10)
top_locations
```

```python
top_locations.plot(kind='bar', color='teal')

plt.title('Locations with the Highest Concentration of Top-Rated Restaurants')
plt.xlabel('Location')
plt.ylabel('Number of Top-Rated Restaurants')
plt.xticks(rotation=45)
plt.show()
```

####Q8: Are there any seasonal trends in the reviews' sentiments?


```python
# Function to calculate sentiment
def calculate_sentiment(text):
    return TextBlob(text).sentiment.polarity  # Returns a value between -1 and 1

# Apply sentiment analysis to the 'Review' column
resturants_data['Sentiment'] = resturants_data['Review'].apply(calculate_sentiment)


# Extract month from 'Dates'
resturants_data['Month'] = resturants_data['Dates'].dt.month

# Group by month and calculate average sentiment
monthly_sentiment = resturants_data.groupby('Month')['Sentiment'].mean()
```

```python
import matplotlib.pyplot as plt
import calendar

# Convert month numbers to month names for better readability
monthly_sentiment.index = monthly_sentiment.index.map(lambda x: calendar.month_abbr[x])

# Plotting
monthly_sentiment.plot(kind='line', marker='o', color='blue')

plt.title('Average Monthly Sentiment of Reviews')
plt.xlabel('Month')
plt.ylabel('Average Sentiment')
plt.grid(True)
plt.show()
```

We can easily conclude that Customers were more postive in November , Jan and more negeative in May.

####Q10: Is there a correlation between the length of a review and the rating given?

```python
# Calculate the length of each review (number of words)
resturants_data['Review_Length'] = resturants_data['Review'].apply(lambda x: len(x.split()))
correlation = resturants_data[['Review_Length', 'Rating']].corr()
correlation
```

```python
sns.scatterplot(data=resturants_data, x='Review_Length', y='Rating', alpha=0.6)

plt.title('Correlation between Review Length and Rating')
plt.xlabel('Review Length (Number of Words)')
plt.ylabel('Rating')
plt.show()
```

While there is a correlation, the value of -0.205 is quite low, indicating that other factors are in determining the rating that is not captured solely by the length of the review.

 Also, from the above we can see satisfied customers leave shorter reviews, often just summarizing their experience with a positive note, which could explain why not all high ratings have long reviews.Customers who have had negative experiences might be more verbose in their reviews, detailing their dissatisfaction.
